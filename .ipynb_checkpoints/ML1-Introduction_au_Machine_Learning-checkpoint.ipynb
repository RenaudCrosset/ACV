***Expliquez en termes simples comment fonctionne un algorithme ML :***  
Le Machine Learning peut être défini comme étant une technologie d’intelligence artificielle qui permet de créer des modèles permettant aux machines d’apprendre sans avoir été au préalable explicitement programmées à cet effet.    
<br>On distingue différents types d’algorithmes de Machine Learning. Généralement, ils peuvent être répartis en trois catégories :
<br>
- les algorithmes supervisés (données étiquetées)
    - les algorithmes de classification (prédictions non-numériques)
        - k-nearest neighbors (kNN, weighted kNN)
        - naive bayes classifier (binary independance naive bayes, multinominal naive bayes)
        - discriminant analysis
    - les algorithmes de régression (prédictions numériques)
        - régression linéaire
        - lasso regression (L1 penalty, square penalty)
        - ridge regression (L2 penalty, absolute value penalty)
        - neural networks regression
    - les deux (classification et regression)
        - arbre de décision
        - forêt aléatoire
        - machine à vecteur de support (SVM, SVM multiple classes, SVM multiple kernels)
        - bootstrap aggregating
- les algorithmes non supervisés (données non étiquetées)
    - les algorithmes de clustering (pour trouver des patterns ou groupes d’objets similaires)
        - ANN (Artificial Neural Networks)
        - k-means clustering
        - mean-shift clustering
        - Density Based Spatial Clustering of Applications with Noise (DBSCAN clustering)
        - hierarchical agglomerative clustering
        - gaussian mixture modelling (GMM)
    - les algorithmes d’association (pour trouver des liens entre des objets)
        - apriori algorithm
        - FP-Growth algorithm
    - les algorithmes de réduction dimensionnelle (pour choisir ou extraire des caractéristiques)
        - principal component analysis (PCA)
        - canonical correlation analysis
- l'apprentissage par renforcement
    - Q-learning
    - R-learning
    - TD learning

    
>- **arbre de décision** (decision tree) : on utilise un ensemble de données pour lesquelles on connaît la valeur de la variable-cible afin de construire l'arbre (données dites étiquetées), puis on extrapole les résultats à l'ensemble des données de test. Les arbres de décision font partie des algorithmes les plus populaires en apprentissage automatique.
       
>- **forêt aléatoire** (random forest) : une forêt aléatoire est un groupe d'arbres de décision, travaillant de manière indépendante sur une vision d'un problème. La théorie veut qu'un grand nombre d'arbres non corrélés créé des prédictions plus précises qu'un arbre de décision individuel, chacun produisant une estimation, et c'est l'assemblage des arbres de décision et de leurs analyses, qui va donner une estimation globale, le volume d'arbres contribuant à protéger des erreurs individuelles et du surajustement. Cet algorithme combine les concepts de sous-espaces aléatoires et de bagging.
    
>- **Gradient Boosting** : le renforcement du gradient : il s’agit d’un ensemble de “weak learners”, créés les uns après les autres, formant un strong learner. Chaque “weak learner” est entraîné pour corriger les erreurs des “weak learners” précédents (par rapport au Bagging, les algorithmes ne sont plus indépendants). Néanmoins, les “weak learners” ont tous autant de poids dans le système de votation, peu importe leur performance (contrairement à Adaboost).
    
>- **machine à vecteur de support** (ou séparateurs à vaste marge) : ce sont des classificateurs qui reposent sur deux idées clés (la notion de marge maximale et la notion de fonction noyau), qui permettent de traiter des problèmes de discrimination non linéaire, et de reformuler le problème de classement comme un problème d'optimisation quadratique.

````python
df_clean = df.copy()
print(df_clean['ELEM_POINT_ID'].count())
df_clean.index = df_clean.ELEM_POINT_ID
````

````python
# mise à la première colonne de notre colonne d'intérêt

def col_prems(col_name, df):
    '''
    retourne une liste réarrangée pour mettre en première place notre colonne d'intérêt
    '''
    cols = df.columns.tolist()
    cols.remove(col_name)
    cols.insert(0, col_name)
    return cols

df_clean = df_clean.reindex(columns=col_prems('ANNEEDEPLANTATION', df_clean))
````

## gestion des colonnes avec informations redondantes

````python
def test_col_redondantes(df):
    '''
    Pour trouver les colonnes redondantes (ID et description, etc)
    '''
    cols = df.columns.tolist()
    idx=0
    i=0
    liste=[]
    liste1=[]
    for col1 in cols:
        i+=1
        for col2 in cols[i:]:
            df['test']= df[col2].astype('string') + df[col1].astype('string')
            if df[col1].dropna().count() == df[col2].dropna().count() and \
               df[col1].nunique() == df[col2].nunique() and \
               df[col1].nunique() == df['test'].nunique():
                cols_tested=[col1, col2]
                found=False
                for cols_red in liste:
                    if list(filter(lambda x: x in cols_red, cols_tested))!=[]:
                        cols_red.extend(cols_tested)
                        found=True                     
                if found == False:
                    liste.append(cols_tested)
                
    if 'test' in df.columns:
        del df['test']
    
    for i, elem in enumerate(liste):
        liste[i]=sorted(list(set(elem)))
    
    return liste

# print(len(set(zip(df_clean['SOUS_CATEGORIE'], df_clean['SOUS_CATEGORIE_DESC']))))

test_col_redondantes(df_clean)
````

````python
cols_to_drop=[]
[cols_to_drop.extend(col1) for *col1, col2  in test_col_redondantes(df_clean)]
cols_to_drop
````

````python
df_clean.drop(columns=cols_to_drop, inplace=True)
````

## traitement données Geo

````python
#Récupérer les coordonnées sous forme de dictionnaires (un par row)
dict_coord = df_clean['GeoJSON'].map(json.loads) #json.loads() is a function to convert a Python string to a dictionary

# Séparer longitude et latitude
lon = dict_coord.apply(lambda x: x['coordinates'][0])
lat = dict_coord.apply(lambda x: x['coordinates'][1])

# Ajouter les colonnes lon et lat à notre df, et enlever la colonne GeoJSON
df_clean['LATITUDE'] = lat
df_clean['LONGITUDE'] = lon
df_clean.drop('GeoJSON', axis=1, inplace=True)

fig, ax = plt.subplots(figsize=(12,12))
#sns.scatterplot(data=df_clean, y='lat', x='lon', s=1, ax=ax)
sns.scatterplot(data=df_clean, y='LATITUDE', x='LONGITUDE', hue='ADR_SECTEUR', ax=ax)
````

````python
def no_info_col(df, col1):
    '''
    Pour trouver les colonnes systématiquement non renseignées 
    lorsque notre colonne d'intérêt n'est pas renseignée.
    '''
    filtered_df = df[(df['ANNEEDEPLANTATION'].isna()==True) | (df['ANNEEDEPLANTATION'].isnull()==True)]
    
    cols = filtered_df.columns.tolist()
    liste=[]
    for col2 in cols:
        if col2!=col1:
            if filtered_df[col2].nunique()==0:
                liste.append(col2)
    return liste

print(cols_to_drop:=no_info_col(df_clean, 'ANNEEDEPLANTATION'))
````

#### model build - linear regression

build linear regresion module and estimate test error using cross validation.

<div class="alert alert-block alert-info">

[**Cross-Validation**](https://www.youtube.com/watch?v=fSytzGwwBVw) : 
With data, we have to :
- first estimate the parameters for the machine learning methods (train, 75% of the data for training)
- second evaluate how well the machine learning methods work (test, 25% of the data for testing)  
    
>**cross-validation** allow to alternatively choose another block of data for testing. In the end, every block of data is used for testing and we can compare methods by seeing how well they performed.

- vocabulary :
    - k blocks => **k-fold cross-validation**  
    - extreme case : each row is a block => **leave one out cross-validation**
    - **tuning parameter** : cross-validation could help find the best value.</div>

````python
lin_reg = LinearRegression()

scores = cross_val_score(lin_reg, X_trainT, y_train, scoring='neg_mean_squared_error', cv=3)
scores

pd.Series(np.sqrt(-scores)).describe()
````

# Step 6 : Fine-tune your models and combine them into a great solution

## Cross validation

We should use CV to get a better estimate of what will happen on testing set (we could be overfitting if we only fit and evaluate on the train).

We can also use it to choose the best hyperparameters for promising models....this is what we do now!

### Tuning models

Forest looks promising. Lets search to find better hyper parameters with CV.

[sklearn.ensemble.RandomForestRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html)

[sklearn.model_selection.GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

````python
param_grid = [
    # try two different searches
    {'n_estimators': [3, 10, 30, 100], 'max_features': [2, 4, 6, 8, 'auto']},

    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor()#random_state=42)
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=False)
grid_search.fit(X_trainT, y_train)
````

````python
results = pd.DataFrame(grid_search.cv_results_)
results[results.rank_test_score == 1]
````

````python
np.sqrt(-results[results.rank_test_score == 1][['mean_test_score']])
````

This is a slight improvement. To ensure that our CV evaluations are correct lets take a look at the test set with this final model we have found! (finally)

<a id="step7"> </a>
# Step 7 : Present your solution

If we choose this as our final model (and it's performance is good enough to put into production!) then we do a final evaluation on the test set to assure that it is likely to perform well once we move to put it into production.

````python
estimator = grid_search.best_estimator_
````

````python
X_testT = full_pipeline.transform(X_test)

tuned_forest_preds = estimator.predict(X_testT)
tuned_forest_mse = mean_squared_error(y_test, tuned_forest_preds)

print('Root mean squared error: ', np.sqrt(tuned_forest_mse))
````

Conclusion: This is not far from our CV evaluation so we conclude that this is approximately how we expect our model to perform in the real world. 

The final model is much better model than the naive and linear models. Although finally depends on the users (tree keepers of Grenoble) whether it is useful or not!

It is possible we could use some of the other X data that was removed or not included, or transform futher/combine some of these X variables to improve the model further.
